```python
# Data Preparation
data = data.drop("LastName",axis=1)
data = data.drop("ExitDate",axis=1)
data = data.drop("ADEmail",axis=1)
data = data.drop("BusinessUnit",axis=1)
data = data.drop("TerminationType",axis=1)
data = data.drop("TerminationDescription",axis=1)
data = data.drop("EmployeeType",axis=1)
data = data.drop("Division",axis=1)
data.info()
```

```python
# Pembersihan Data
def clean_data(column_names):

    characters_to_remove = ["$", "%", "&"]

    for column_name in column_names:
        for char in characters_to_remove:

            data["Division"] = data["Division"].str.replace(char, "")
```

```python
# Data Manipulation & Transformation
def convert_and_divide(column_name, value):

    data[column_name] = data[column_name].astype(float)
    data[column_name] = data[column_name] / value
```
